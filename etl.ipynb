```python
import pandas as pd
import gzip
import json
import numpy as np
import ast
```

```python
def parse(path):
    with gzip.open(path, 'rt', encoding='utf-8') as g:
        for line in g:
            yield json.loads(line)

rows = []

# Loop through the objects of the JSON parse and append each row to the list
for line in parse('steam_games.json.gz'):
    row = pd.json_normalize(line)
    rows.append(row)
```

```python
#concateno las filas en un unico DataFrame
df = pd.concat(rows, ignore_index=True)
```
